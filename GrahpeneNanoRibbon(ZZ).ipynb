## Graphene Nano-Ribbon with vacants.

```python
!pip install kwant
from math import pi, sqrt, tanh
from matplotlib import pyplot as plt
import kwant
import numpy as np

# define both lattices and sublattices
graphene = kwant.lattice.honeycomb()   # modified lattice
graphene0 = kwant.lattice.honeycomb()  # pure lattice; from now on 0 at the end means no vacants.
a0, b0 = graphene0.sublattices
a, b = graphene.sublattices
```

```python
## defining a lattice w/o vacants for comparison ##
syst0= kwant.Builder()
def make_system0(l = 10, w = 10, pot = 0.1):
    def nrb(pos):
        (x,y) = pos
        return 0 <= x <= l and 0.5 <= y <=w
    def potential(site):                        
        (x,y) = site.pos
        return pot    
    syst0[graphene0.shape(nrb, (0, 0))] = potential 
    hoppings = ()
    syst0[[kwant.builder.HoppingKind(*hopping) for hopping in hoppings]] = -1
    syst0[graphene.neighbors()] = -1
    sym = kwant.TranslationalSymmetry((-1,0))      
    sym.add_site_family(a0, other_vectors= [(-1,2)]) 
    sym.add_site_family(b0, other_vectors= [(-1,2)]) 
    lead = kwant.Builder(sym)
    def lead_shape(pos):
        (x,y) = pos
        return 0.5 <= y <= w
    lead[graphene0.shape(lead_shape, (1,1))] = pot
    lead[graphene0.neighbors()] = -1
    return syst0, [lead, lead.reversed()]
```

```python
## defining a lattice w vacants ##
# syst == system with vacants #
syst = kwant.Builder()
def make_system(l = 10, w = 10, pot = 0.1):
    def nrb(pos):
        (x,y) = pos
        return 0 <= x <= l and 0.5 <= y <=w
# onsite energy of TB syst
    def potential(site):                        
        (x,y) = site.pos
        return pot    
    syst[graphene.shape(nrb, (0, 0))] = potential 
# hoppings
    hoppings = ()
    syst[[kwant.builder.HoppingKind(*hopping) for hopping in hoppings]] = -1
    syst[graphene.neighbors()] = -1
# vacants
    # deleting one site in lattice notation a(i,j), b(i,j) for a,b sublattices
    del syst[a(1,2)]
    # by shape
    # def in_hole(site):
    #    x, y = site.pos  
    #    return 2 < x < 3 and 2 < y < 4

    # for site in filter(in_hole, list(syst.sites())):
    #    del syst[site]
# symmetries
    sym = kwant.TranslationalSymmetry((-1,0))       # traditional ((x,y) vectors)
    sym.add_site_family(a, other_vectors= [(-1,2)]) # (*) append (accepts lattice vectors)
    sym.add_site_family(b, other_vectors= [(-1,2)]) # (*) append (accepts lattice vectors)
    
    # (*) i append a fundamental unity cell for defining a periodicity for t.sym
    # w/ other_vectors which form a base, i create a way of giving directions fot t.sym
    # which respect the periodicity of the lattice (needed for TranslationalSymmetry function)
    # which cannot be defined by lattice vectors.

# leads
    lead = kwant.Builder(sym)
    def lead_shape(pos):
        (x,y) = pos
        return 0.5 <= y <= w
    lead[graphene.shape(lead_shape, (1,1))] = pot
    lead[graphene.neighbors()] = -1
    return syst, [lead,lead.reversed()]   
```

```python
# system generation, checkpoint code
syst0, leads = make_system0(l= 20, w = 7, pot = 0.1)
syst, leads = make_system(l= 20, w = 7, pot = 0.1)
kwant.plot(syst); 
kwant.plot(syst0);
```

```python
def plot_conductance(syst, energies):
    # transmission as function of energy                                
    data = []
    data0 = []
    for energy in energies:
        smatrix = kwant.smatrix(syst, energy)    
        data.append(smatrix.transmission(0, 1))
    for energy in energies:
        smatrix0 = kwant.smatrix(syst0, energy)
        data0.append(smatrix0.transmission(1,0))
    # here smatrix.transmission computes the transmission (or reflexion)
    # as transmission(lead_out, lead_in) so it computes the transmission between
    # lead 1 and 0. smatrix is computed with wave function approach.

    plt.plot(energies, data, 'b', label = 'w vacants')
    plt.plot(energies, data0, 'k', label = 'w/o vacants')
    plt.xlabel("energy [t]")
    plt.ylabel("conductance [2e^2/h]")
    plt.title('Conductance')
    plt.legend()
    plt.show()
# bandstructure

def plot_bandstructure(flead, momenta):
# kwant.phy... devuelve un array de NumPy array que contiene los autobvalores de todos los
# modos para un momento
    bands = kwant.physics.Bands(flead)
    energies = [bands(k) for k in momenta]

    plt.figure()
    plt.plot(momenta, energies)
    plt.xlabel("momenta [(lattice constant)^-1]")
    plt.ylabel("energy [t]")
    plt.show()   
```

```python
## Plotting
# finishing the systems

for lead in leads:
    syst.attach_lead(lead)
syst = syst.finalized()

for lead in leads:
    syst0.attach_lead(lead)
syst0 = syst0.finalized()

# Plotting the system
kwant.plot(syst);
# kwant.plot(syst0); for ribbon w/o vacants
# Plotting band structure of lead 0 (left)
momenta = [-pi + 0.02 * pi * i for i in range(101)]
plot_bandstructure(syst.leads[0], momenta);
# Plotting conductance
plot_conductance(syst, energies = np.linspace(0, 1.3))
```
